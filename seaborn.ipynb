We use distrubution plots to visualize the distribution of quantitative data.

Let's discuss some plot that allow us to visualize the distribution of a data set. These plots are :
1. distplot/histplot
2. jointplot
3. pairplot.
4. rugplot

**Distrubution Plots**

```python
import seaborn as sns #importing new library seaborn
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique() #check the different different values of that column
```

```python
sns.histplot(df['total_bill'], bins = 20, kde = 'true') #dislpot or histplot both are same and result are also same
#bins means how much bars you want by default it is 20
#kde meaning kernal density estimator how our data is normal or skewed
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins = 20, kde = 'true')
plt.subplot(1,2,2)
sns.histplot(df['tip'], bins = 20, kde = 'true')
#this is how you can make more than one plot in one canvas
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'scatter') #joint plot can use 2 column at once
#kind means select which type of pattern you need
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'kde')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'reg')
```

```python
sns.pairplot(df, hue = 'sex') #compare numeric value to each other
#hue is use to define the data like male and female
```

```python
sns.pairplot(df, hue = 'size', palette='rainbow') #you can use multiple color with palette
```

```python
sns.rugplot(df) #where a value exist this plot is use for only this
```

**Categorical Plot**

Categorical Data plots
There are a few main plot types for this:
1. boxplot
2. violinplot
3. stripplot
4. swarmplot
5. barplot
6. countplot

```python
sns.countplot(x = 'sex', data = df,) #apply count plot on gender
```

```python
sns.countplot(x = 'sex', data = df, hue = df['smoker']) # in one time you can also add another data like hue
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = df) #we have to put continues value on y-axis
```

you can change the estimator object to your own function, that converts a vector to a scalar
mean (default) : np.mean
median : np.median
sum = np.sum
count : len(to count the number of observation)
standard deviation : np.std
Minimum : np.min
Maximum : np.max

```python
import numpy as np
sns.barplot(x = df['sex'], y=df['total_bill'],estimator=np.sum)
```

```python
sns.barplot(x = df['sex'], y=df['tip'],estimator=np.sum)
```

```python
sns.boxplot(x='tip', y='day', data = df, palette='rainbow') #this is how you can use box plot and that dot showing in image that is outliers
```

**Box plot and Violinplot**

boxplots and violinplots are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparision between vriables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are detemined to be "outliers" using a method that is a function of the inter-quantile range.


```python
sns.violinplot(x='day', y='total_bill', data = df, palette='rainbow')
```

**Stripplot and Swarmpot**

The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drwan on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observation along with some representation of the underlaying distribution.

The swarmplot is similar to stripplot(), but the points are adjusted (only along categorical axis) so that they don't overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observation (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

```python
sns.stripplot(x='day', y='tip', data = df)
```

```python
sns.swarmplot(x='day', y='tip', data=df)
```

```python
sns.violinplot(x='tip', y='day', data= df)
sns.swarmplot(x='tip', y='day', data = df) #this is how you can show two plots in one canvas
```
