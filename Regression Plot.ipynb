# lmplot()

we will see how it is created and what are the values we can put inside it

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips) #this is how you can use lmplot
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex') #this is how you can use hue
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm') #this is how you can add colours
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm', markers=['o','v']) #this is how you can change markers
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm', markers=['o','v'], scatter_kws={'s':80}) #this is how you can set size of markers
```
