Object Oriented Way

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch)

```python
fig = plt.figure() #to know the size of canvas or white background
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.2,0.5,0.3,0.4])
axis2.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,5),dpi=100,facecolor='red',edgecolor='blue') #here you can set the canvas size and dpi means quality of your canvas
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.2,0.5,0.3,0.4])
axis2.plot(x,y)
```
